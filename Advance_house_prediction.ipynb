```
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
  print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values ')
```

SINCE they are many missing values , we need to find the relationshp betwwen values and sales price

lets plot some diagram for this realtionship

```
for feature in features_with_na:
  data = dataset.copy()
  # lets make a vaarible that indicates 1 if the observation was missing or else zero
# because we want to build countplot
  data[feature] = np.where(data[feature].isnull(),1,0)
  # lets calculate mean sales price whr the information is missing or pre
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```
# analyze the temporal datetime varibles
# we will check wheter thre relationship betwwen yer the house is solf
#and year 
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('median house price')
plt.title('house price vs Yearsold')
```

from this we difine that dis not good but ww know year increses price also increses


```
# here we will compare diff bertween all yeaars feeatures with saesprice
for feature in year_feature:
  if feature != 'YrSold':
    data = dataset.copy()
    # we will capture the differnce betwwen yer varible and yer h
    data[feature]=data['YrSold']-data[feature]

    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```
# numerical varibles 2 type 1)discrete 2)continous
discrete_feature = [feature for feature in numerical_features if 
                    len(dataset[feature].unique())<25 and 
                        feature not in year_feature+['Id']]
print('discrete varibles count : {}'.format(len(discrete_feature)))
```

```
# lets find out relationship between than saleprice
for feature in discrete_feature:
  data=dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Saleprice')
  plt.title(feature)
  plt.show()
```

```
# numerical varibles 2 type 1)discrete 2)continous
continous_feature = [feature for feature in numerical_features
                     if feature not in discrete_feature
                     + year_feature+['Id']]
print('continous features count : {}'.format(len(continous_feature)))
```

```
# lets analyze cont. varib;es
for feature in continous_feature:
  data = dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel("count")
  plt.title(feature)
  plt.show()
```

from these we can see the distributon is not normally distributed

to overcome this we using logrithims of transfomation.

```

for feature in continous_feature:
 data = dataset.copy()
 if 0 in data[feature].unique():
   pass
 else:
  data[feature] = np.log(data[feature])
     # for sALES PRICE LOG --->
  data['SalePrice']=np.log(data['SalePrice'])
  plt.scatter(data[feature],data['SalePrice'])
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
      
```
